```python
#a)
#Your code:
sqtf_sum = df['sqft_living'] + df['sqft_lot'] + df['sqft_above'] + df['sqft_basement']

fig0=plt.figure()
plt.scatter(df['sqft_living'],df['price'], marker='.')
plt.xlabel('square foot')
plt.ylabel('price')
plt.show()

fig1 = plt.figure()
plt.scatter(sqtf_sum, df['price'], marker='.')
plt.xlabel('square foot sum')
plt.ylabel('price')
plt.show()
#Hint plt.scatter
```

## Exercise 3
a) Create boxplot of apartment's grade versus it's price. Compare with plot 2b).  
b) Create mosaic plot of any two categorical variables.

```python
#a)
#Your code:
data = df.pivot(columns='grade',values='price')
grouped_data = [df[df['grade'] == grade]['price'] for grade in sorted(df['grade'].unique())]
# Tworzenie wykresu pudełkowego dla wszystkich ocen na jednym obrazku
plt.figure(figsize=(10, 6))
plt.boxplot(grouped_data)

# Dodanie tytułu i etykiet
plt.xlabel('Grade')
plt.ylabel('Price')

# Wyświetlenie wykresu
plt.show()
#Hint: plt.boxplot
```

> ***Odpowiedź:*** Tutaj mamy więcej informacji, wykresy pudełkowe dają nam informację o kwartylach 1/4 oraz 3/4, kreska pomarańczowa wskazuje nam mediane, a w kwadracie znajduję się 50% danych. Kropki nad wykresem obrazują nam dane, które odbiegają swoją wartością od reszty. Kreski to tzw. "whiskers", wskazują one najmniejszą oraz najwiekszą wartość pomijając dane, które odbiegają, czyli bez tzw "outlinerów". Aczkolwiek, na wykresach pudełkowych nie możemy zobaczyć każdej jednej zależności, wiec w tym sensie są mniej dokładne

```python
#b)
from statsmodels.graphics.mosaicplot import mosaic

data = {
    'grade': df['grade'],
    'condition': df['condition']
}

#Your code:
mosaic(pd.DataFrame(data), ['grade','condition'])
plt.xlabel('grade')
plt.ylabel('condition')
plt.figure(figsize=(20, 22))
plt.show()
```

> *Komentarz:* Widzimy, że w tego typu wykresach, dane są przedstawione w formie bardziej przystępnej jeśli chodzi o ich wizualizacje, jest to lepsze rozwiązanie niż histogram(w tym przypadku), ponieważ daję nam ono bardziej kompaktowe spojrzenie na dane tego typu
